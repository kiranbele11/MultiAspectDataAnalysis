### Correlation

```python
a = [20,37,45,56,47]
b = [21,36,44,76,50]
#Covariance mmatrix: Symmetric matrix i.e, transpose of this matrix is same

#using numpy
np.cov(a,b)
```

```python
np.cov(a,a)
```

```python
np.cov(b,b)
```

```python
a1 = np.array([20,37,45,56,47])
b1 = np.array([21,36,44,76,50])
```

```python
#Standard Deviation
np.std(a1)
```
